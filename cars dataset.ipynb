now we are going to use the altair plot to form a plot to show the max cost of a car.
we are going to make this using the altair library


```python
max_cost_car = df.groupby('manufacturer')['cost of car'].max().reset_index()
max_cost_car
```

```python
#to make the scatter plot for the max cost of a car made by a manufacture.
import altair as alt
# we have to dowload this in the pip manager.
```

```python
scatter_plot = alt.Chart(max_cost_car).mark_point().encode(x='manufacturer' ,y='cost of car' )
# this has been made using the altair library for making scatter plot
scatter_plot
```
